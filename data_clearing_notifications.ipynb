```python
import pandas as pd
import numpy as np
```

```python
def transform_date(column_dataframe):
    date = [f'{str(i)[0:4]}'+ '-' +f'{str(i)[4:6]}'+ '-'+ f'{str(i)[6:8]}' for i in column_dataframe]
    return date
```

```python
reactions = pd.read_csv('VigiMed_Reacoes.csv', encoding = 'ANSI', sep = ';')
medications = pd.read_csv('VigiMed_Medicamentos.csv', encoding = 'ANSI', sep = ';', on_bad_lines='skip')
notifications = pd.read_csv('VigiMed_Notificacoes.csv', encoding = 'ANSI', sep = ';')
```

```python
notifications['DATA_ULTIMA_ATUALIZACAO'] = transform_date(notifications['DATA_ULTIMA_ATUALIZACAO'])
notifications['DATA_INCLUSAO_SISTEMA'] = transform_date(notifications['DATA_INCLUSAO_SISTEMA'])
notifications['DATA_NOTIFICACAO'] = transform_date(notifications['DATA_NOTIFICACAO'])
notifications['DATA_NASCIMENTO'] = transform_date(notifications['DATA_NASCIMENTO'])
```

```python
notifications.head(5)
```

```python
notifications.rename(columns = {'#UF': 'UF'}, inplace = True)
```

```python
notifications['TIPO_ENTRADA_VIGIMED'] = notifications['TIPO_ENTRADA_VIGIMED'].str.replace('Pacientes\xa0e Profissionais de Saúde', 'Pacientes e Profissionais de Saúde')
notifications['TIPO_ENTRADA_VIGIMED'] = notifications['TIPO_ENTRADA_VIGIMED'].str.replace('Empresas\xa0Farmacêuticas', 'Empresas e Farmacêuticas')
```

```python
notifications['IDADE_MOMENTO_REACAO'] = notifications['IDADE_MOMENTO_REACAO'].str.replace('nan', 'NaN')
```

```python
idade = []
for i in notifications['IDADE_MOMENTO_REACAO'].fillna('null'):
    i = str(i)
    if 'ano' in i:
        m = 1
    if 'mês' in i:
        m = 12
    if 'dia' in i:
        m = 365
    if 'semana' in i:
        m = 52
    if 'década' in i or i == 'ano' or i == 'dia' or i == 'hora' or i == 'mês':
        i = 'null'
    
    if i != 'null':    
        i = i.replace('ano', '').replace('mês', '').replace('dia', '').replace('semana', '') \
        .replace('década', '').replace('hora', '').replace('seman', '')
        i = float(i)//m
        
    idade.append(i)
```

```python
new_idade= [i if i != 'null' else None for i in idade]
```

```python
v = notifications['IDADE_GESTACIONAL_MOMENTO_REACAO'].fillna('null').values
```

```python
idade_gestacao = []
for i in v:
    i = str(i)
    if 'Semanas' in i:
        m = 1
    if 'Mês' in i:
        m = 30
    if 'Dias' in i:
        m = 7
    if 'Trimestre' in i:
        m = 12
    
    if i != 'null':    
        i = i.replace('Semanas', '').replace('Mês', '').replace('Dias', '').replace('Trimestre', '')
        if m == 1:
            i = float(i) * 1
        if m == 30:
            i = (float(i)*m)//7
        if m == 7:
            i = float(i)/m
        if m == 12:
            i = float(i) * m
        
    idade_gestacao.append(i)
```

```python
notifications['IDADE_GESTACIONAL_MOMENTO_REACAO'] = [int(i) if i != 'null' else None for i in idade_gestacao]
```
